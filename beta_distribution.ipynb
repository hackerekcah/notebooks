```python
%pylab
import scipy
from scipy.stats import beta
```

# symmetric beta
## if alpha > 1, like gaussian, center on 0.5

```python
def symmetric_beta(bt_one):
    x = np.linspace(0.01, 0.99, 100)
    y = []
    if bt_one:
        alphas = [1, 10, 100, 1000]
    else:
        alphas = [0.1, 0.3, 0.5, 0.9]
    for alpha in alphas:
            y.append(beta.pdf(x, alpha, alpha))

    plt.plot(x,np.array(y).T)
    plt.legend([str(a) for a in alphas])
```

```python
symmetric_beta(bt_one=True)
```

```python
symmetric_beta(bt_one=False)
```

# asymmetric beta
## for alpha belong to (0,1), bigger alpha, more smooth

```python
def asymmetric_beta():
    x = np.linspace(0.01, 0.99, 100)
    y = []
    alphas = [0.1, 0.3, 0.5, 0.9]
    for alpha in alphas:
            y.append(beta.pdf(x, alpha+1, alpha))

    plt.plot(x,np.array(y).T)
    plt.legend(['alpha='+str(a) for a in alphas])
```

```python
asymmetric_beta()
```
